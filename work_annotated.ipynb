#### Converting Stringified Lists to Python Lists using `ast.literal_eval()`
We use the built-in `ast` (Abstract Syntax Trees) module to safely parse stringified Python objects (like lists of dictionaries) into actual Python lists.  
This function `convert()` specifically extracts the `'name'` field from each dictionary within the stringified list.

 **Why this is needed:**  
In the dataset, columns like `genres`, `keywords`, `cast`, and `crew` are stored as strings that look like lists of dictionaries.  
To work with them effectively, we first convert them into real Python list objects.


```python
import ast
# ast is a built-in library in Python that allows you to parse and process Python expressions
def convert(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

Taking only the name of the genres

```python
movies['genres'] = movies['genres'].apply(convert) # it will convert the string to list of dictionaries
```


Applying the same above things to keywords too as genre.

```python
movies['keywords'] = movies['keywords'].apply(convert) # it will convert the string to list of dictionaries
```

### Extracting Top 3 Cast Members from the Cast Feature
The `cast` column contains a stringified list of dictionaries, where each dictionary holds information about an actor.

 **Goal:**  
We only want the **top 3 actors** for each movie, and we will extract just their `name` fields for simplicity and relevance.

This helps in:
- Reducing noise from less important cast members.
- Representing each movie with its key actors, which can aid in similarity-based recommendations.


```python
import ast
# ast is a built-in library in Python that allows you to parse and process Python expressions
def convert3(obj):
    L = []
    counter =0
    for i in ast.literal_eval(obj):
        if counter != 3:
            L.append(i['name'])
            counter += 1
        else:
            break
    return L
```

```python
movies['cast'] = movies['cast'].apply(convert3) # it will convert the string to list of dictionaries
```

###  Removing Space from Names to Avoid Recommendation Confusion
Sometimes, different individuals can have similar or partially matching names (e.g., `'Sam Worthington'` vs `'Sam Mendes'`).  
If we treat these names as space-separated tokens, it may lead to incorrect matches or recommendations.

 **Solution:**  
We **remove spaces** from all names (e.g., `'Sam Worthington'` â†’ `'SamWorthington'`) to treat each name as a **unique single token**.

 **Why this is important:**  
This avoids mixing up distinct individuals in vector-based similarity models like TF-IDF or CountVectorizer, ensuring more accurate content-based recommendations.


### Removing Spaces in Genre, Keywords, Cast, and Crew Names for Consistency
To maintain consistency and prevent issues in text vectorization (like TF-IDF or CountVectorizer),  
we remove spaces within individual names or terms across the following columns:
- `genres`
- `keywords`
- `cast`
- `crew`

 **Implementation:**  
Using `lambda` with `.apply()`, we replace `" "` with `""` in each list element.

 **Why this matters:**  
Models treat space-separated words as different tokens.  
For example, `"Science Fiction"` becomes `["Science", "Fiction"]` which is misleading. By removing the space (`"ScienceFiction"`), we preserve its identity as a single genre.

This step ensures better feature representation and avoids ambiguity during vectorization and similarity calculations.


```python
movies['genres'] = movies['genres'].apply(lambda x: [i.replace(" ","")for i in x]) # it will convert the string to list
movies['keywords'] = movies['keywords'].apply(lambda x: [i.replace(" ","")for i in x]) # it will convert the string to list
movies['cast'] = movies['cast'].apply(lambda x: [i.replace(" ","")for i in x]) # it will convert the string to list
movies['crew'] = movies['crew'].apply(lambda x: [i.replace(" ","")for i in x]) # it will convert the string to list
```

###  Importing Libraries
This cell imports the required Python libraries for the project.

```python
from sklearn.feature_extraction.text import CountVectorizer
```

###  Creating CountVectorizer for Text Feature Extraction
We initialize `CountVectorizer` with:
- `max_features=5000` to keep the top 5000 frequent words
- `stop_words='english'` to remove common English words (like "the", "and", etc.)

This converts text data into numerical vectors for similarity calculation.


```python
cv = CountVectorizer(max_features=5000,stop_words='english') # it will create a count vectorizer object with the max features of 5000 and stop words as english
```

### Converting Tags to Numerical Vectors
We apply `CountVectorizer` to the `tags` column to transform the text into numerical feature vectors.

 `cv.fit_transform()` learns the vocabulary and transforms the text into a 2D array of word counts for each movie.


```python
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors[0]
```

### Initializing Porter Stemmer for Word Normalization

### Stemming Words in the 'tags' Column
We define the `stem()` function to reduce words to their root form using the `PorterStemmer`.  
This is applied to the `tags` column to normalize the text data for better similarity calculations.


```python
def stem(text):
    y =[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
new_df['tags'] = new_df['tags'] = new_df['tags'].apply(stem)
```

###  Calculating Cosine Similarity Between Movies
We use `cosine_similarity` from `sklearn` to compute similarity scores between movies based on their `tags` vectors.  
This will help in identifying movies with similar content.


```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

### Using Enumerate to Retain Indices While Sorting
We use `enumerate()` to retain the index of each similarity score, ensuring we don't lose track of movie positions when sorting by similarity.


```python
list(enumerate(similarity[0]))
```

### Sorting Movies by Similarity Score
We sort the movies based on their similarity to the first movie (in descending order) using `enumerate()` to retain the indices.  
The result shows the top 5 most similar movies, excluding the first movie itself.


```python
sorted(list(enumerate(similarity[0])),reverse= True,key = lambda x:x[1])[1:6]
```

### Sorting Movies by Similarity Score
We sort the movies based on their similarity to the first movie (in descending order) using `enumerate()` to retain the indices.  
The result shows the top 5 most similar movies, excluding the first movie itself.


```python
def recommend(movie):
    movie_index = new_df[new_df['title'] == movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)),reverse= True,key = lambda x:x[1])[1:6]

    for i in movies_list:
        print(new_df.iloc[i[0]].title)
    
```

### Recommending Movies Similar to 'Batman Begins'
We call the `recommend()` function with the movie title `'Batman Begins'` to display the top 5 movies similar to it based on cosine similarity.


```python
recommend('Batman Begins')
```
